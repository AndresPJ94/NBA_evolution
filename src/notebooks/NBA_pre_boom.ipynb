```python
#LIBRERIAS
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVC
import pickle
import warnings
```

```python
#ESTADÍSTICAS 2006/07
df_2007 = pd.read_csv('../data/raw/06-07.csv', encoding='ISO-8859-1', sep=",", index_col=0)
df_2007.loc[:, 'Season'] = '2006-2007'

#ELIMINAMOS JUGADORES DUPLICADOS (TRASPASOS)
sin_dupes_2007 = df_2007.drop_duplicates(subset=['Player'], keep='first')
pos_2007 = sin_dupes_2007.loc[::, 'Pos']

df_2007_filtrado = df_2007[(df_2007['MP'].between(15,45)) & (df_2007['G'].between(40,82))].reset_index(drop=True)

#ESTADÍSTICAS 2007/08
df_2008 = pd.read_csv('../data/raw/07-08.csv', encoding='ISO-8859-1', sep=",", index_col=0)
df_2008.loc[:, 'Season'] = '2007-2008'

#ELIMINAMOS JUGADORES DUPLICADOS (TRASPASOS)
sin_dupes_2008 = df_2008.drop_duplicates(subset=['Player'], keep='first')
pos_2008 = sin_dupes_2008.loc[::, 'Pos']

df_2008_filtrado = df_2008[(df_2008['MP'].between(15,45)) & (df_2008['G'].between(40,82))].reset_index(drop=True)

#ESTADÍSTICAS 2008/09
df_2009 = pd.read_csv('../data/raw/08-09.csv', encoding='ISO-8859-1', sep=",", index_col=0)
df_2009.loc[:, 'Season'] = '2008-2009'

#ELIMINAMOS JUGADORES DUPLICADOS (TRASPASOS)
sin_dupes_2009 = df_2009.drop_duplicates(subset=['Player'], keep='first')
pos_2009 = sin_dupes_2009.loc[::, 'Pos']

df_2009_filtrado = df_2009[(df_2009['MP'].between(15,45)) & (df_2009['G'].between(40,82))].reset_index(drop=True)

#ESTADÍSTICAS 2009/10
df_2010 = pd.read_csv('../data/raw/09-10.csv', encoding='ISO-8859-1', sep=",", index_col=0)
df_2010.loc[:, 'Season'] = '2009-2010'

#ELIMINAMOS JUGADORES DUPLICADOS (TRASPASOS)
sin_dupes_2010 = df_2010.drop_duplicates(subset=['Player'], keep='first')
pos_2010 = sin_dupes_2010.loc[::, 'Pos']

df_2010_filtrado = df_2010[(df_2010['MP'].between(15,45)) & (df_2010['G'].between(40,82))].reset_index(drop=True)

#ESTADÍSTICAS 2010/11
df_2011 = pd.read_csv('../data/raw/10-11.csv', encoding='ISO-8859-1', sep=",", index_col=0)
df_2011.loc[:, 'Season'] = '2010-2011'

#ELIMINAMOS JUGADORES DUPLICADOS (TRASPASOS)
sin_dupes_2011 = df_2011.drop_duplicates(subset=['Player'], keep='first')
pos_2011 = sin_dupes_2011.loc[::, 'Pos']

df_2011_filtrado = df_2011[(df_2011['MP'].between(15,45)) & (df_2011['G'].between(40,82))].reset_index(drop=True)

#CREAMOS EL DATAFRAME GLOBAL
total = pd.concat([df_2007_filtrado, df_2008_filtrado, df_2009_filtrado, df_2010_filtrado, df_2011_filtrado]).drop(['Player-additional'], axis=1).reset_index(drop=True)

#GUARDAMOS EN UN NUEVO CSV LOS DATOS YA FILTRADOS
total.to_csv('../data/processed/2000s_filtrado.csv')  


#Cinco categorías (sin filtrar por partidos ni min jugados)
cinco_cat = total[total['Pos'].isin(["SG", "PF", "C", "SF", "PG"])]
cinco_cat.reset_index()

#Outliers (sin filtrar)
outliers= total[total['Pos'].isin(["PG-SG", 'SF-SG', 'PF-SF', 'SG-PG', 'PF-C', 'SF-PF', 'SG-PF', 'C-PF', 'SG-SF'])]

#Creamos una nueva columna llamada Binario donde dividimos a los jugadores entre posiciones exteriores e interiores.
cinco_cat.loc[cinco_cat['Pos'] == 'C', 'Binario'] = 'Interior'
cinco_cat.loc[cinco_cat['Pos'] == 'PF', 'Binario'] = 'Interior'
cinco_cat.loc[cinco_cat['Pos'] == 'SG', 'Binario'] = 'Exterior'
cinco_cat.loc[cinco_cat['Pos'] == 'SF', 'Binario'] = 'Exterior'
cinco_cat.loc[cinco_cat['Pos'] == 'PG', 'Binario'] = 'Exterior'

definitivas = cinco_cat.iloc[::, 7:29].drop(['FG%', '3P%', '2P%', 'eFG%', 'FT%', 'TRB'], axis=1)

outliers['Pos'].value_counts()
len(df_2007.columns)
```

```python
#Realizamos un grid search para cinco categorías con una Reg Log

#Dividimos en train, test y definimos X e y.
X = cinco_cat.iloc[::, 7:29].drop(['FG%', '3P%', '2P%', 'eFG%', 'FT%', 'TRB'], axis=1)
y = cinco_cat['Pos']
x_train, x_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.30,
                                                    random_state = 101)


#Escalamos las variables
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)

#Definimos los parámetros de búsqueda en la regresión logística
logisticRegr = LogisticRegression()

parameters = {
    'solver': ['newton-cg', 'liblinear', 'lbfgs'],
    'penalty': ['l1', 'l2', 'elasticnet']
}

grid = GridSearchCV(estimator = logisticRegr,
                   param_grid = parameters,
                    n_jobs = -1,
                    scoring = 'accuracy',
                    cv = 10,
                    verbose = 3)


#Entenamos el modelo
grid.fit(x_train_scaled, y_train)

#Sacamos el mejor score y los mejores parámetros
print("Best estimator:", grid.best_estimator_)
print("Best params:", grid.best_params_)
print("Best score:", grid.best_score_)
print('Prediccion Test: ',grid.best_estimator_.score(x_test_scaled, y_test))
```

```python
#Realizamos un grid search para DOS categorías con una Log Reg

#Dividimos en train, test y definimos X e y.
X = cinco_cat.iloc[::, 7:29].drop(['FG%', '3P%', '2P%', 'eFG%', 'FT%', 'TRB'], axis=1)
y = cinco_cat['Binario']
x_train, x_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.30,
                                                    random_state = 101)

#Escalamos las variables

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)

#Definimos los parámetros de búsqueda en la regresión logística

logisticRegr = LogisticRegression()

parameters = {
    'solver': ['newton-cg', 'liblinear', 'sag', 'saga', 'lbfgs'],
    'penalty': ['l1', 'l2', 'elasticnet']
}

grid = GridSearchCV(estimator = logisticRegr,
                   param_grid = parameters,
                    n_jobs = -1,
                    scoring = 'accuracy',
                    cv = 10)


#Entenamos el modelo
grid.fit(x_train_scaled, y_train)


#Sacamos el mejor score y los mejores parámetros
print("Best estimator:", grid.best_estimator_)
print("Best params:", grid.best_params_)
print("Best score:", grid.best_score_)
print('Prediccion Test: ',grid.best_estimator_.score(x_test_scaled, y_test))
```

```python
#Realizamos un grid search para SVM con DOS CATEGORÍAS

#Definimos X e y
X = cinco_cat.iloc[::, 7:29].drop(['FG%', '3P%', '2P%', 'eFG%', 'FT%', 'TRB'], axis=1)
y = cinco_cat['Binario']

x_train, x_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.30,
                                                    random_state = 42)


#Escalamos las variables
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)

#Definimos el grid de búsqueda
params_grid = {
    'kernel': ['linear', 'rbf', 'sigmoid'],
    'C': [0.001, 0.01, 0.5, 1, 5, 10, 15, 20, 100],
    'gamma': ['scale', 'auto'],
    'coef0': [-10, -1, 0, 0.1, 0.5, 1, 10, 100]
}

svm_model = GridSearchCV(SVC(), params_grid, cv=10)
svm_model.fit(x_train_scaled, y_train)


#Vemos los parámetros óptimos para el grid search con SVM
print('Best score for training data:', svm_model.best_score_,"\n") 
print('Best C:',svm_model.best_estimator_.C,"\n") 
print('Best Kernel:',svm_model.best_estimator_.kernel,"\n")
print('Best Gamma:',svm_model.best_estimator_.gamma,"\n")
print('Best Coef0:',svm_model.best_estimator_.coef0,"\n")
print('Prediccion Test: ',svm_model.best_estimator_.score(x_test_scaled, y_test))
```

```python
#Realizamos un grid search para SVM con CINCO CATEGORÍAS

#Definimos X e y
X = cinco_cat.iloc[::, 7:29].drop(['FG%', '3P%', '2P%', 'eFG%', 'FT%', 'TRB'], axis=1)
y = cinco_cat['Pos']

x_train, x_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.30,
                                                    random_state = 42)


#Escalamos las variables
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)

#Definimos el grid de búsqueda
params_grid = {
    'kernel': ['linear', 'rbf', 'sigmoid'],
    'C': [0.001, 0.01, 0.5, 1, 5, 10, 15, 20, 100],
    'gamma': ['scale', 'auto'],
    'coef0': [-10, -1, 0, 0.1, 0.5, 1, 10, 100]
}

svm_model = GridSearchCV(SVC(), params_grid, cv=10)
svm_model.fit(x_train_scaled, y_train)


#Vemos los parámetros óptimos para el grid search con SVM
print('Best score for training data:', svm_model.best_score_,"\n") 
print('Best C:',svm_model.best_estimator_.C,"\n") 
print('Best Kernel:',svm_model.best_estimator_.kernel,"\n")
print('Best Gamma:',svm_model.best_estimator_.gamma,"\n")
print('Best Coef0:',svm_model.best_estimator_.coef0,"\n")
print('Prediccion Test: ',svm_model.best_estimator_.score(x_test_scaled, y_test))
```

```python
#Sacamos la matriz de confusión para comparar ambas épocas
cm_labels = np.unique(y)
predictions = svm_model.predict(x_test_scaled)
cm_array = confusion_matrix(y_test, predictions)
cm_array_df = pd.DataFrame(cm_array, index=cm_labels, columns=cm_labels)
sns.heatmap(cm_array_df, annot=True, annot_kws={"size": 12}, cmap='rocket_r', cbar=False);
plt.title('NBA años 2000')
plt.savefig('../img/confusion_pre_boom.jpg')

#Predicciones de los jugadores con categorías mixtas
Predicha = pd.Series(svm_model.predict(scaler.fit_transform(outliers.iloc[::, 7:29].drop(['FG%', '3P%', '2P%', 'eFG%', 'FT%', 'TRB'], axis=1))))
Pos = pd.Series(outliers['Pos']).reset_index(drop=True)
Jugador = pd.Series(outliers['Player']).reset_index(drop=True)
Temporada = pd.Series(outliers['Season']).reset_index(drop=True)

df = pd.concat([Jugador, Temporada, Predicha, Pos], axis=1).reset_index(drop=True)
df.columns = ['Jugador', 'Temporada', 'Predicha', 'Posición']
df.iloc[17:32]


#Ahora guardamos el modelo para poder reutilizarlo en un futuro
with open('../modelos/NBA_pre_boom.model', "wb") as archivo_salida:
    pickle.dump(svm_model.best_estimator_, archivo_salida)
```

```python
#Decission Tree para 5 categorías (con Baggins)

#Definimos X e y
X = cinco_cat.iloc[::, 7:29].drop(['FG%', '3P%', '2P%', 'eFG%', 'FT%', 'TRB'], axis=1)
y = cinco_cat['Pos']

x_train, x_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.30,
                                                    random_state = 101)


#Escalamos las variables
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)

kfold = model_selection.KFold(n_splits=10)
dtc = DecisionTreeClassifier()

model = BaggingClassifier(base_estimator=dtc,
                n_estimators = 100,
                random_state = 101)

bagging = model
print("Accuracy Train: ",model_selection.cross_val_score(model, x_train_scaled, y_train, cv=kfold).mean())
print("Accuracy Test: ",model_selection.cross_val_score(model, x_test_scaled, y_test, cv=kfold).mean())
```
